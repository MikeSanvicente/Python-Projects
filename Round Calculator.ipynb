```python
"""
Customizable Round Calculator
-----------------------------------
Skills: 
1. Conditional Statements
2. Functions
3. Input Handling
4. List Operations
5. Loops

"""

def round_calculator():
        """
        Customizable Round Calculator
        
        This function allows users to calculate the winner of a customizable number
        of rounds fight based on round scores between two fighters. It handles input
        validation for the number of rounds, fighter names, and round scores, and
        provides clear instructions and error messages to the user.
        """

        # Initialize lists to store scores
        score_f1 = []
        score_f2 = []
        rd = 1

        # Get the number of rounds from the user
        while True:
            num_rounds = input("Enter the number of rounds: ")
            try:
                if int(num_rounds) in range(1,13):
                    num_rounds=int(num_rounds)
                    print("The fight is scheduled for",num_rounds,"rounds")
                    break
                else:
                    print("Invalid input. Please enter a number between 1 and 12.")
            except:
                print("Invalid input. Please enter a number between 1 and 12.")
        
        # Get the names of the fighters
        while True:
            f1_name = input("Who's the first fighter?: ").strip()
            f2_name = input("Who's the second fighter?: ").strip()
            
            if f1_name and f2_name:
                break  
            else:
                print("Both fighter names are required. Please enter both fighter names.")
            
        print(f1_name, "vs", f2_name, "scorecard:")
        
        # Loop through rounds
        while rd <= num_rounds:
            while True:
                score1 = input(f"{f1_name}'s {rd} round score is: ")
                score2 = input(f"{f2_name}'s {rd} round score is: ")
                try:
                    score1=int(score1)
                    score2=int(score2)
                    if score1>=0 and score1 <=11 and score2>=0 and score2 <=10:
                        break
                    else:
                        print(f"Please enter a valid number:\n      1 to 10 for round score \n      11 for a {f1_name} win by KO \n      0 for a {f2_name} win by KO")
                except ValueError:
                    print("Invalid input. Please enter a valid numeric score.")
                
        # If fighter 1 won by knockout (score1 == 11) 
        # If fighter 2 won by knockout (score1 == 0)

            if score1 == 0:
                rd += num_rounds
                print(f2_name, "won by KO!")
                break
            elif score1 > 10:
                rd += num_rounds
                print(f1_name, "won by KO!")
                break
            else:
                # Append round scores to respective lists
                score_f1.append(score1)
                score_f2.append(score2)

                print(f"Round {rd}\n{f1_name} {score1} {f2_name} {score2}")
                rd += 1

        # Calculate the total scores for each fighter
        result1 = sum(score_f1)
        result2 = sum(score_f2)

        # Display final scores and the winner
        print(f1_name, "'s final score is:", result1, ",", f2_name, "'s final score is:", result2)

        if result1 > result2:
            print(f1_name, "won by", result1 - result2, "points.")
        elif result1 < result2:
            print(f2_name, "won by", result2 - result1, "points.")
        else:
            print("It's a tie!")

# Call the function to execute it
round_calculator()
```
